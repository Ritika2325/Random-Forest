```python
# Bagged Decision Trees for Classification
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
fraud=pd.read_csv("C:/Fraudcheck.csv")
```

```python
fraud.head()
```

```python
fraud=pd.get_dummies(fraud,columns=['Undergrad','Marital.Status','Urban'])
```

```python
fraud.head()
```

```python
fraud["income"]="<30000"
fraud.loc[fraud["Taxable.Income"]>=30000,"income"]="Good"
fraud.loc[fraud["Taxable.Income"]<=30000,"income"]="Risky"
```

```python
fraud.head()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder=preprocessing.LabelEncoder()
fraud['income']=labelencoder.fit_transform(fraud['income'])
```

```python
fraud.head()
```

```python
x=fraud.iloc[:,1:10]
x
```

```python
y=fraud.iloc[:,10]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
kfold = KFold(n_splits=10, random_state=7)
cart = DecisionTreeClassifier()
num_trees = 100
```

```python
model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=7)

results = cross_val_score(model, x, y, cv=kfold)

X=results.mean()
X
```

```python
# Random Forest Classification
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
num_trees = 10
max_features = 3

kfold = KFold(n_splits=10, random_state=7)

model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)

results = cross_val_score(model, x, y, cv=kfold)

print(results.mean())
```

```python
results
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
#Tunning the model using GridsearchCV
n_estimators = [int(x) for x in np.linspace(start=10 , stop=80 , num=10)]
n_jobs = [int(y) for y in np.linspace(start=2 , stop=8 )]
oob_score= [True , False]
criterion=["gini", "entropy"]
```

```python
param_grid = {'n_estimators' : n_estimators,
            'n_jobs' : n_jobs,
             'oob_score' : oob_score,
             'criterion' : criterion}
```

```python
RFmodel = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_Grid = GridSearchCV(estimator = RFmodel , param_grid = param_grid , cv=3 )
```

```python
rf_Grid.fit(x_train,y_train)
```

```python
rf_Grid.best_params_
```

```python
print(f'trainAcc - : {rf_Grid.score(x_train,y_train):.3f}')
print(f'testAcc - : {rf_Grid.score(x_test,y_test):.3f}')
```

#### Univariate Feature Selection

```python
# Feature Extraction with Univariate Statistical Tests (Chi-squared for classification)
from pandas import read_csv
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
array = fraud.values
X = array[:,1:10]
Y = array[:,10]

# feature extraction
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y)

# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)
```

#### Recursive Feature Elimination

```python
# Feature Extraction with RFE
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

array = fraud.values
X = array[:,1:10]
Y = array[:,10]

# feature extraction

model = LogisticRegression(max_iter=500)
rfe = RFE(model, 4)
fit = rfe.fit(X, Y)
```

```python
#Num Features: 
fit.n_features_
```

```python
#Selected features
fit.support_
```

```python
# Feature Ranking:
fit.ranking_
```
